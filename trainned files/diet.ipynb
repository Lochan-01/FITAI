```python
import pandas as pd

diet_df = pd.read_csv("dataset1.csv")

diet_df.head()
```

```python
print(diet_df.columns.tolist())
```

```python
diet_df = diet_df[
    [
        "Name",
        "Calories",
        "ProteinContent",
        "CarbohydrateContent",
        "FatContent",
        "FiberContent",
        "RecipeIngredientParts"
    ]
]

diet_df.head()
```

```python
diet_df = diet_df.dropna()

diet_df = diet_df.reset_index(drop=True)

print(diet_df.shape)
```

```python
def assign_goal(calories):

    if calories <= 300:
        return "Weight Loss"

    elif calories <= 600:
        return "Maintain"

    else:
        return "Muscle Gain"


diet_df["Goal"] = diet_df["Calories"].apply(assign_goal)
```

```python
def meal_type(calories):

    if calories < 250:
        return "Snack"

    elif calories < 450:
        return "Breakfast"

    elif calories < 700:
        return "Lunch"

    else:
        return "Dinner"


diet_df["Meal"] = diet_df["Calories"].apply(meal_type)
```
